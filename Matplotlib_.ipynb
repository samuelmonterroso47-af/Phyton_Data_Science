**9 - Has recibido un desafío para calcular la raíz cuadrada de una lista de números, identificando cuáles resultan en un número entero. La lista es la siguiente:**
`numeros = [2, 8, 15, 23, 91, 112, 256]`

```python
import math

numeros = [2, 8, 15, 23, 91, 112, 256]

print("Calculando la raíz cuadrada de cada número:")
for num in numeros:
    raiz_cuadrada = math.sqrt(num)
    # Comprueba si la raíz cuadrada es un número entero
    if raiz_cuadrada == int(raiz_cuadrada):
        print(f"La raíz cuadrada de {num} es {int(raiz_cuadrada)} (entero).")
    else:
        print(f"La raíz cuadrada de {num} es {raiz_cuadrada:.2f} (no entero)." )
```

**10 - Haz un programa para una tienda que vende césped para jardines. Esta tienda trabaja con jardines circulares y el precio del metro cuadrado de césped es de R$ 25,00. Pide a la persona usuaria el radio del área circular y devuelve el valor en reales de cuánto tendrá que pagar.**

```python
import math

PRECIO_POR_METRO_CUADRADO = 25.00 # R$

# Solicita el radio del área circular
radio = float(input("Ingresa el radio del jardín circular (en metros): "))

# Calcula el área del círculo (Área = π * radio^2)
area = math.pi * (radio ** 2)

# Calcula el costo total
costo_total = area * PRECIO_POR_METRO_CUADRADO

print(f"El área del jardín es: {area:.2f} metros cuadrados")
print(f"El valor total a pagar por el césped es: R$ {costo_total:.2f}")
```
