1 - Haz un programa que solicite a la persona usuaria ingresar dos números decimales y calcular la división entre estos números. El código debe incluir un manejo de error, indicando el tipo de error que se generó si la división no es posible.

Prueba el programa con el segundo valor numérico de la entrada igual a 0. También prueba usando caracteres textuales en la entrada para verificar los tipos de errores que ocurren.

```python
def dividir_numeros():
    try:
        # Solicitar al usuario que ingrese el primer número decimal
        num1_str = input("Ingrese el primer número decimal: ")
        num1 = float(num1_str)

        # Solicitar al usuario que ingrese el segundo número decimal
        num2_str = input("Ingrese el segundo número decimal: ")
        num2 = float(num2_str)

        # Realizar la división
        resultado = num1 / num2
        print(f"El resultado de la división es: {resultado}")

    except ValueError:
        # Manejo de error si la entrada no es un número decimal válido
        print("Error: Entrada inválida. Asegúrese de ingresar solo números decimales.")
    except ZeroDivisionError:
        # Manejo de error si el segundo número es cero
        print("Error: No se puede dividir por cero.")
    except Exception as e:
        # Manejo de cualquier otro error inesperado
        print(f"Se ha producido un error inesperado: {e}")

# --- Pruebas del programa ---

print("\n--- Prueba 1: División válida ---")
dividir_numeros()

print("\n--- Prueba 2: Segundo número igual a 0 (ZeroDivisionError) ---")
dividir_numeros()

print("\n--- Prueba 3: Caracteres textuales en la entrada (ValueError) ---")
dividir_numeros()
```

2 - Haz un programa que solicite a la persona usuaria ingresar un texto que será una clave a buscar en el siguiente diccionario: edades = {'Júlia': 16, 'Carol': 23, 'Alberto': 19, 'Roberta': 17}, almacenando el resultado del valor en una variable. El código debe incluir un manejo de error KeyError, imprimiendo la información 'Nombre no encontrado' en caso de error, e imprimir el valor si no ocurre ninguno.

Prueba el programa con un nombre presente en una de las claves del diccionario y con uno que no esté en el diccionario para verificar el mensaje de error.

```python
edades = {'Júlia': 16, 'Carol': 23, 'Alberto': 19, 'Roberta': 17}

def buscar_edad():
    nombre_a_buscar = input("Ingrese el nombre de la persona a buscar: ")
    try:
        edad_encontrada = edades[nombre_a_buscar]
        print(f"La edad de {nombre_a_buscar} es: {edad_encontrada}")
    except KeyError:
        print("Nombre no encontrado")

print("\n--- Prueba 1: Nombre presente en el diccionario ---")
buscar_edad()

print("\n--- Prueba 2: Nombre no presente en el diccionario ---")
buscar_edad()
```

3 - Crea una función que reciba una lista como parámetro y convierta todos los valores de la lista a flotantes. La función debe incluir un manejo de error indicando el tipo de error generado y devolver la lista si no ha ocurrido ningún error. Por último, debe tener la cláusula finally para imprimir el texto: 'Fin de la ejecución de la función'.

```python
def convertir_a_flotantes(lista):
    lista_flotantes = []
    try:
        for elemento in lista:
            try:
                lista_flotantes.append(float(elemento))
            except ValueError:
                print(f"Error de conversión: No se pudo convertir '{elemento}' a flotante. (ValueError)")
                # Optionally re-raise the error or handle it differently if the function should fail on first error
                return None # Devolvemos None si hay un error en la conversión de algún elemento
            except TypeError:
                print(f"Error de tipo: El elemento '{elemento}' no puede ser convertido a flotante. (TypeError)")
                return None # Devolvemos None si hay un error de tipo
        print("La lista ha sido convertida exitosamente a flotantes.")
        return lista_flotantes
    except Exception as e:
        print(f"Se ha producido un error inesperado durante la ejecución: {e}")
        return None
    finally:
        print('Fin de la ejecución de la función.')

# --- Pruebas de la función ---

print("\n--- Prueba 1: Lista con todos los elementos válidos ---")
lista_valida = [1, 2.5, '3', '4.0']
resultado1 = convertir_a_flotantes(lista_valida)
if resultado1 is not None:
    print(f"Resultado: {resultado1}")

print("\n--- Prueba 2: Lista con un elemento no convertible (ValueError) ---")
lista_invalida_valor = [10, 'abc', 20.5]
resultado2 = convertir_a_flotantes(lista_invalida_valor)
if resultado2 is not None:
    print(f"Resultado: {resultado2}")

print("\n--- Prueba 3: Lista con un elemento de tipo incorrecto (TypeError) ---")
lista_invalida_tipo = [1, [2], 3.0]
resultado3 = convertir_a_flotantes(lista_invalida_tipo)
if resultado3 is not None:
    print(f"Resultado: {resultado3}")
```

4 - Crea una función que reciba dos listas como parámetros y agrupe los elementos uno a uno de las listas, formando una lista de tuplas de 3 elementos. El primer y segundo elemento de la tupla son los valores en la posición i de las listas y el tercer elemento es la suma de los valores en la posición i de las listas.

```python
def agrupar_listas_y_sumar(lista1, lista2):
    # Verificar si las listas tienen la misma longitud
    if len(lista1) != len(lista2):
        raise IndexError('La cantidad de elementos en cada lista es diferente.')

    # Inicializamos una lista vacía para almacenar las tuplas resultantes
    lista_de_tuplas = []

    # Usamos zip para iterar sobre ambas listas simultáneamente
    for i, (elemento1, elemento2) in enumerate(zip(lista1, lista2)):
        try:
            # Calculamos la suma de los elementos
            suma = elemento1 + elemento2
            # Creamos la tupla de 3 elementos
            tupla = (elemento1, elemento2, suma)
            # Añadimos la tupla a nuestra lista de resultados
            lista_de_tuplas.append(tupla)
        except TypeError:
            # Manejo de error si los elementos no son sumables
            print(f"Error: Los elementos en la posición {i} ('{elemento1}', '{elemento2}') no se pueden sumar. Omitiendo esta tupla.")
        except Exception as e:
            # Manejo de cualquier otro error inesperado
            print(f"Se ha producido un error inesperado en la posición {i}: {e}")

    return lista_de_tuplas

# --- Nuevas Pruebas de la función ---

print("\n--- Prueba 1: Valores sin error ---")
lista1_valida = [4, 6, 7, 9, 10]
lista2_valida = [-4, 6, 8, 7, 9]
try:
    resultado1 = agrupar_listas_y_sumar(lista1_valida, lista2_valida)
    print(f"Resultado: {resultado1}")
except IndexError as e:
    print(f"Error: {e}")

print("\n--- Prueba 2: Listas con tamaños diferentes (IndexError) ---")
lista1_diferente_largo = [4, 6, 7, 9, 10, 4]
lista2_diferente_largo = [-4, 6, 8, 7, 9]
try:
    resultado2 = agrupar_listas_y_sumar(lista1_diferente_largo, lista2_diferente_largo)
    print(f"Resultado: {resultado2}")
except IndexError as e:
    print(f"Error: {e}")

print("\n--- Prueba 3: Listas con valores incoherentes (TypeError) ---")
lista1_incoherente = [4, 6, 7, 9, 'A']
lista2_incoherente = [-4, 'E', 8, 7, 9]
try:
    resultado3 = agrupar_listas_y_sumar(lista1_incoherente, lista2_incoherente)
    print(f"Resultado: {resultado3}")
except IndexError as e:
    print(f"Error: {e}")
```

5 - Como desafío, se te ha asignado la tarea de desarrollar un código que contabiliza las puntuaciones de estudiantes de una institución educativa de acuerdo con sus respuestas en una prueba. Este código debe ser probado para un ejemplo de 3 estudiantes con una lista de listas en la que cada lista tiene las respuestas de 5 preguntas objetivas de cada estudiante. Cada pregunta vale un punto y las alternativas posibles son A, B, C o D.

Si alguna alternativa en una de las pruebas no está entre las alternativas posibles, debes lanzar un ValueError con el mensaje "La alternativa [alternativa] no es una opción de alternativa válida". El cálculo de las 3 notas solo se realizará mediante las entradas con las alternativas A, B, C o D en todas las pruebas. Si no se lanza la excepción, se mostrará una lista con las notas en cada prueba.

Datos para la prueba del código:

Respuestas de la prueba:

respuestas = ['D', 'A', 'B', 'C', 'A']
Copia el código
A continuación, hay 2 listas de listas que puedes usar como prueba:

Notas sin excepción:

tests_sin_ex = [['D', 'A', 'B', 'C', 'A'], ['C', 'A', 'A', 'C', 'A'], ['D', 'B', 'A', 'C', 'A']]
Copia el código
Notas con excepción:

tests_con_ex = [['D', 'A', 'B', 'C', 'A'], ['C', 'A', 'A', 'E', 'A'], ['D', 'B', 'A', 'C', 'A']]


```python
def contabilizar_puntuaciones(respuestas_correctas, tests_estudiantes):
    notas = []
    alternativas_validas = {'A', 'B', 'C', 'D'}

    for i, test_estudiante in enumerate(tests_estudiantes):
        nota_actual = 0
        if len(test_estudiante) != len(respuestas_correctas):
            raise ValueError(f"El test del estudiante {i+1} tiene un número incorrecto de preguntas. Esperadas: {len(respuestas_correctas)}, Recibidas: {len(test_estudiante)}.")

        for j, respuesta_estudiante in enumerate(test_estudiante):
            # Validar que la alternativa sea una de las permitidas
            if respuesta_estudiante not in alternativas_validas:
                raise ValueError(f"La alternativa '{respuesta_estudiante}' en la pregunta {j+1} del estudiante {i+1} no es una opción de alternativa válida.")

            # Si la alternativa es válida, comparar con las respuestas correctas
            if respuesta_estudiante == respuestas_correctas[j]:
                nota_actual += 1
        notas.append(nota_actual)

    return notas

# Datos para la prueba del código
respuestas = ['D', 'A', 'B', 'C', 'A']

# Notas sin excepción:
tests_sin_ex = [['D', 'A', 'B', 'C', 'A'], ['C', 'A', 'A', 'C', 'A'], ['D', 'B', 'A', 'C', 'A']]

# Notas con excepción:
tests_con_ex = [['D', 'A', 'B', 'C', 'A'], ['C', 'A', 'A', 'E', 'A'], ['D', 'B', 'A', 'C', 'A']]

print("\n--- Prueba con 'tests_sin_ex' (sin excepción) ---")
try:
    puntuaciones_sin_ex = contabilizar_puntuaciones(respuestas, tests_sin_ex)
    print(f"Notas de los estudiantes (sin excepción): {puntuaciones_sin_ex}")
except ValueError as e:
    print(f"Error inesperado: {e}")

print("\n--- Prueba con 'tests_con_ex' (con excepción) ---")
try:
    puntuaciones_con_ex = contabilizar_puntuaciones(respuestas, tests_con_ex)
    print(f"Notas de los estudiantes (con excepción): {puntuaciones_con_ex}")
except ValueError as e:
    print(f"Error esperado: {e}")
```

6 - Estás trabajando con procesamiento de lenguaje natural (NLP) y, en esta ocasión, tu líder te pidió que crees un fragmento de código que reciba una lista con las palabras separadas de una frase generada por ChatGPT.

Necesitas crear una función que evalúe cada palabra de este texto y verifique si el tratamiento para quitar los símbolos de puntuación (',', '.', '!' y '?') se realizó. De lo contrario, se lanzará una excepción del tipo ValueError señalando el primer caso en que se detectó el uso de una puntuación a través de la frase "El texto presenta puntuaciones en la palabra "[palabra]"". Esta solicitud se centra en el análisis del patrón de frases generadas por la inteligencia artificial.

Datos para probar el código:

Lista tratada:

lista_tratada = ['Python', 'es', 'un', 'lenguaje', 'de', 'programación', 'poderoso', 'versátil',
                  'y', 'fácil', 'de', 'aprender', 'utilizado', 'en', 'diversos', 'campos', 'desde',
                  'análisis', 'de', 'datos', 'hasta', 'inteligencia', 'artificial']
Copia el código
Lista no tratada:

lista_no_tratada = ['Python', 'es', 'un', 'lenguaje', 'de', 'programación', 'poderoso,', 'versátil',
                  'y', 'fácil,', 'de', 'aprender', 'utilizado', 'en', 'diversos', 'campos,', 'desde',
                  'análisis', 'de', 'datos', 'hasta', 'inteligencia', 'artificial!']


```python
def verificar_puntuacion_en_texto(lista_palabras):
    puntuaciones = {',', '.', '!', '?'}

    for palabra in lista_palabras:
        for caracter in palabra:
            if caracter in puntuaciones:
                raise ValueError(f"El texto presenta puntuaciones en la palabra \"{palabra}\".")

    # Si no se encuentra ninguna puntuación, no se lanza ninguna excepción
    print("El texto no presenta puntuaciones. El tratamiento se realizó correctamente.")
    return True

# Datos para probar el código:

# Lista tratada:
lista_tratada = ['Python', 'es', 'un', 'lenguaje', 'de', 'programación', 'poderoso', 'versátil',
                  'y', 'fácil', 'de', 'aprender', 'utilizado', 'en', 'diversos', 'campos', 'desde',
                  'análisis', 'de', 'datos', 'hasta', 'inteligencia', 'artificial']

# Lista no tratada:
lista_no_tratada = ['Python', 'es', 'un', 'lenguaje', 'de', 'programación', 'poderoso,', 'versátil',
                  'y', 'fácil,', 'de', 'aprender', 'utilizado', 'en', 'diversos', 'campos,', 'desde',
                  'análisis', 'de', 'datos', 'hasta', 'inteligencia', 'artificial!']

print("\n--- Prueba con lista_tratada (sin excepción) ---")
try:
    verificar_puntuacion_en_texto(lista_tratada)
except ValueError as e:
    print(f"Error: {e}")

print("\n--- Prueba con lista_no_tratada (con excepción) ---")
try:
    verificar_puntuacion_en_texto(lista_no_tratada)
except ValueError as e:
    print(f"Error: {e}")
```
